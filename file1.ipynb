```python
for i in glob.glob('C:/Users/chandra.gorripat/Desktop/refine data/*[0-9][0-9][0-9]'):
    json_files = glob.glob(os.path.join(i, "*.json"))
    csv_files = glob.glob(os.path.join(i, "*.csv"))
    print(i)
    for f in json_files:
        #print(f)
        t=f[:-4]+'csv'
        #print(t)
        df1=pd.read_csv(t)
        #print(df1)
        with open(f,'r') as f:
         data = json.loads(f.read())
         # Flatten data
         df2 = pd.json_normalize(data)
        #*********************************************************************************
        df2['min_sum']=df1['sum'].min()
        df2['max_sum']=df1['sum'].max()
        df2['mean_sum']=df1['sum'].mean()
        df2['std_sum']=df1['sum'].std()
        df2['duration_of_event']=df1['timestamp'].max()
       # df1['sum_dev']=((((df1['sum']-df1['sum'].mean())/df1['sum'].mean())*100)//1)*1
        ########################################
        
        df1['new_sum']=(df1['sum']//10)*10
        arr1=df1['new_sum'].unique()
        element1,count1=arr1[0],df1[df1['new_sum']==arr1[0]].shape[0]
        for i in range(1,len(arr1)):
              tc1=df1[df1['new_sum']==arr1[i]].shape[0]
              if tc1>count1:
                element1=arr1[i]
                count1=tc1 
        df2['max_10_dev']=element1 
        df2['max_10_dur']=count1/40
        #df1['sum_dev']=((((df1['sum']-df1['sum'].mean())/df1['sum'].mean())*100)//20)*20
        df=df.append(df2)
        #print(df2) 
```

```python
for i in range(df.shape[0]):
    df['Event_start_time'][i]=df['Event_start_time'][i][2:-6]
    df['Event_start_time'][i]=df['Event_start_time'][i].replace('-','/')
    df['Event_start_time'][i]=df['Event_start_time'][i].replace('T',' ')
    
```

```python
for i in range(df.shape[0]):
        datetime_str =df['Event_start_time'][i] 
        df['Event_start_time'][i] = datetime.strptime(datetime_str, '%y/%m/%d %H:%M:%S')
```

```python
df['Total_Tare_weight']=0
```

```python
for i in range(df.shape[0]):
    df['Total_Tare_weight'][i]=df['Tare_weight'][i][0]+df['Tare_weight'][i][1]+df['Tare_weight'][i][2]+df['Tare_weight'][i][3]
```

```python
df['hour_day']=0
df['month']=np.nan
for i in range(df.shape[0]):
    df['hour_day'][i]=df['Event_start_time'][i].hour#+round(df['Event_start_time'][i].minute/60,1)
    df['month'][i]=df['Event_start_time'][i].month
    
```

```python
df['urination']=0
df['defecation']=0
df['elimination']=0
df['non_elimination']=1
for i in range(df.shape[0]):
    if 'urination' in df['Tags'][i]:
        df['urination'][i]=1
        df['elimination'][i]=1
    if 'defecation' in df['Tags'][i]:
        df['defecation'][i]=1
        df['elimination'][i]=1  
    if df['elimination'][i]==1:
        df['non_elimination'][i]=0
```
