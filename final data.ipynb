```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os
import json
from datetime import datetime
```

```python
df1=pd.read_csv('file1.csv')
df2=pd.read_csv('file2.csv')
```

```python
df1
```

```python

df1.drop(['Unnamed: 0','Free_text','Tags','Name_of_cat','Weight_of_cat','min_sum','max_sum','mean_sum','std_sum','duration_of_event'],axis=1,inplace=True)
```

```python
df2
```

```python

df2.drop(['Unnamed: 0','Device_ID','Event_start_time','Free_text','Tare_weight'],axis=1,inplace=True)
```

```python
df=pd.merge(df1,df2, how ='left', on ='Event_ID')
```

```python
df
```

```python
for i in range(df.shape[0]):
    df['Event_start_time'][i]=df['Event_start_time'][i][2:]
    df['Event_start_time'][i]=df['Event_start_time'][i].replace('-','/')
    df['Event_start_time'][i]=df['Event_start_time'][i].replace('T',' ')
for i in range(df.shape[0]):
        datetime_str =df['Event_start_time'][i] 
        df['Event_start_time'][i] = datetime.strptime(datetime_str, '%y/%m/%d %H:%M:%S')    
```

```python
df['ratio_10_weight']=(df['Total_Tare_weight']-df['max_10_dev'])/df['Weight_of_cat']
df['ratio_min_weight']=(df['Total_Tare_weight']-df['min_sum'])/df['Weight_of_cat']
df['ratio_min_10_weight']=(df['max_10_dev']-df['min_sum'])/df['Weight_of_cat']
```

```python
df['cat_in_box']=0
for i in range(df.shape[0]):
    if 'cat in box' in df['Tags'][i]:
        df['cat_in_box'][i]=1
```

```python
df[df['Weight_of_cat']==0]
```

```python
df['max_10_dur']=df['max_10_dur']/df['duration_of_event']
df['count_max_shift']=df['count_max_shift']/df['duration_of_event']
df['count_min_shift']=df['count_min_shift']/df['duration_of_event']
```

```python

df['month']=0
for i in range(df.shape[0]):
    df['month'][i]=df['Event_start_time'][i].month
    
```

```python
df1=df
```

```python
df['elimination']=0
for i in range(df.shape[0]):
    if 'urination' in df['Tags'][i]:
        df['elimination'][i]=1
    if 'defecation' in df['Tags'][i]:
        df['elimination'][i]=1   
```

```python
df.drop(index=2823,inplace=True)
df.drop(index=2950,inplace=True)
df.drop(index=10997,inplace=True)
```
